```python
customer['Dt_Customer'] = pd.to_datetime(customer['Dt_Customer'])

customer['Dt_Customer'].dtypes
```

```python
# The mathematical equation is sum of number of months between enrolment
# year to 2015 AND one month less to the enrolment date months data.


# ( Assuming we are currently in 2015 January)
customer['Month_Customer'] = ( 2015 - customer['Dt_Customer'].dt.year ) * 12 - ( customer['Dt_Customer'].dt.month - 1 )
```

```python
customer.head()
```

```python
customer["TotalSpendings"] = customer['MntWines'] + customer['MntFruits'] + customer['MntMeatProducts'] + \
                                customer['MntFishProducts'] + customer['MntSweetProducts'] + customer['MntGoldProds']


# forward slash is just to divide the code into multiple lines
```

```python
# On the basis of Age let's divide the customers into different age groups and create a column "AgeGroup". The logic for that is;

#  Age Group is Teen for age less than 19
#  Age Group is Adults for age between 20 and 39
#  Age Group is Middle Age Adults for age between 40 and 59
#  Age Group is Senior for age more than 60


# **Option 1 :**

#    Create an empty column and then fill the volumes
# customer['AgeGroup'] = np.nan


# **option 2 :**

customer.loc[  customer['Age'] <= 19 ,   "AgeGroup"] = 'Teen'   # Making a new column named "AgeGroup"

customer.loc[  (customer['Age'] >=20) & (customer['Age'] <= 39) ,   "AgeGroup"] = 'Adults'
customer.loc[  (customer['Age'] >= 40) & (customer['Age'] <= 59) ,   "AgeGroup"] = 'Middle Age Adults'
customer.loc[  customer['Age'] >= 60 ,   "AgeGroup"] = 'Senior'


```

```python
# Bivariate Analysis on Maritial Status


## Creating Group
grp = customer.groupby("Marital_Status").agg( totalRecords = ("Marital_Status", np.size) , 
                                            TotalSpending = ("TotalSpendings", np.sum) ,
                                            AvgSpending = ("TotalSpendings", np.mean) , 
                                            AvgAge = ("Age", np.mean) )
print( grp )



plt.figure( figsize = (20, 5) )

plt.subplot(1, 4, 1)
plt.pie( x = grp.totalRecords.values , labels = grp.index.values, explode = [0, 0, 0.2, 0, 0, 0, 0, 0] )
plt.title(" Total Records ")

plt.subplot(1, 4, 2)
plt.pie( x = grp.TotalSpending.values , labels = grp.index.values, explode = [0, 0, 0.2, 0, 0, 0, 0, 0] )
plt.title(" Total Spendings ")

plt.subplot(1, 4, 3)
plt.pie( x = grp.AvgSpending.values , labels = grp.index.values, explode = [0, 0, 0.2, 0, 0, 0, 0, 0] )
plt.title(" Avg Spendings ")

plt.subplot(1, 4, 4)
plt.pie( x = grp.AvgAge.values , labels = grp.index.values, explode = [0, 0, 0.2, 0, 0, 0, 0, 0] )
plt.title(" Avg Age ")




plt.show()
```

```python
# Bivariate Analysis on Maritial Status


## Creating Group
grp = customer.groupby("Marital_Status").agg( totalRecords = ("Marital_Status", np.size) , 
                                            TotalSpending = ("TotalSpendings", np.sum) ,
                                            AvgSpending = ("TotalSpendings", np.mean) , 
                                            AvgAge = ("Age", np.mean) )
print( grp )



plt.figure( figsize = (20, 5) )

plt.subplot(1, 4, 1)
grp = grp.sort_values("totalRecords", ascending = False)
plt.bar( height = grp.totalRecords.values , x = grp.index.values , color = "blue")
plt.title(" Total Records ")
plt.xticks(rotation = 45)

plt.subplot(1, 4, 2)
grp = grp.sort_values("TotalSpending", ascending = False)
plt.bar( height = grp.TotalSpending.values , x = grp.index.values , color = "orange")
plt.title(" Total Spendings ")
plt.xticks(rotation = 45)

plt.subplot(1, 4, 3)
grp = grp.sort_values("AvgSpending", ascending = False)
plt.bar( height = grp.AvgSpending.values , x = grp.index.values , color = "green")
plt.title(" Avg Spendings ")
plt.xticks(rotation = 45)

plt.subplot(1, 4, 4)
grp = grp.sort_values("AvgAge", ascending = False)
plt.bar( height = grp.AvgAge.values , x = grp.index.values , color = "pink" )
plt.title(" Avg Age ")
plt.xticks(rotation = 45)




plt.show()
```
